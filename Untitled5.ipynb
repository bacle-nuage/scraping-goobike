```
#####
# library
#####
import requests
from bs4 import BeautifulSoup
import json
import re
import time
```

```
#####
# cost
#####
makers = [
         'maker-harley_davidson'
]
base_url = 'https://www.goobike.com'
```

```
#####
# Functions
#####

# Beautiful Soup
def get_bs(url):
    # スクレイピング対象の URL にリクエストを送り HTML を取得する
    res = requests.get(url)

    # レスポンスの HTML から BeautifulSoup オブジェクトを作る
    # res.encoding = res.apparent_encoding # なぜか文字コードを判定できなかったため直接指定
    res.encoding = 'EUC-JP'
    soup = BeautifulSoup(res.text.encode('utf-8'), 'html.parser')
    # print('res.encoding after start')
    # print(res.encoding)
    # print('res.encoding end')
    # print(res.text)

    time.sleep(1)

    return soup

def get_new_data(maker, cat_types):
  prefix = '/catalog/'
  maker_name = maker.replace('maker-', '').upper()

  result = []
  for c in cat_types:
    request_url = base_url + prefix + maker_name + '/' + c.upper() + '/index.html'
    # request_url = 'https://www.goobike.com/catalog/HARLEY_DAVIDSON/HARLEY_DAVIDSON_EL/index.html'
    print('request url ::' + request_url)

    soup = get_bs(request_url)

    # print(soup.find('ul', attrs={'class': 'year_fee_lists'}))
    if soup.find('ul', attrs={'class': 'year_fee_lists'}).find('li', attrs={'class' : 'year_fee_list'}) is None:
      years = ''
      price = ''
    else:
      years = soup.find('ul', attrs={'class': 'year_fee_lists'}).find('li', attrs={'class' : 'year_fee_list'}).find('dl').find('dd').find('span').get_text()
      price = soup.find('ul', attrs={'class': 'year_fee_lists'}).find('li', attrs={'class' : 'year_fee_list'}).find('dl').find('dd').find('b').get_text()
    print('put data :: , car_type = ' + c + ' years = ' + str(years) + ' , price = ' + str(price))
    
    d = {
        'car_type' : c,\
        'years' : years,\
        'price' : price\
    }

    result.append(d.copy())

  return result

# htmlタグの削除
def del_tag(tag, tag_name):
  needle = '<'+tag_name+'>(.*?)</'+tag_name+'>'
  return re.search(needle, str(tag)).group(1)

# ファイルに書き込み
import csv
def output(file_name, data):
  with open(file_name, "a") as f:
      writer = csv.writer(f, lineterminator='\n')
      # writer.writerow(data[0].keys())
      print(data)
      for i in data:
          writer.writerow(i.values())
```
